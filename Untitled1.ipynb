```python
"""
Task-1:
=======
Create a list of 10 integers (your choice).
Print the first, last, and middle elements.
Print the list in reverse using slicing.
"""
mylist = [2,3,5,8,7,9,1,0,4,6]
print(mylist)
print('First element in my list is', mylist[0])
print('Last element in my list is', mylist[-1])
x1=len(mylist) // 2 - 1
x2=len(mylist) // 2
print('Middle elements of list are:', mylist[x1], 'and', mylist[x2])

#Print list in reverse using slicing
print('Reverse list is', mylist[ ::-1])
```
